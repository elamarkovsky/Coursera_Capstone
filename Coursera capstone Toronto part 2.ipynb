```python
import pandas as pd
```

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", index_col=None)
df=df[0]
df.head(5)
```

Drop rows where Borough is not assigned

```python
df=df[df["Borough"]!="Not assigned"]
df.head(10)
```

Change Neighborhood value to Borough if Neighborhood is not assigned

```python
df["Neighbourhood"].replace("Not assigned", df["Borough"], inplace=True)
df.head(10)
```

Combine neughborhoods with the same postcode

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df.head(25)
```

Merge the two columns

```python
df_com=pd.merge(df, df_coor)
```
